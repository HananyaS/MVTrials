```python
import pandas as pd
import plotly.express as px
```

```python
dataset = 'parkinson'
results_to_optim = 'partial'
```

```python
df = pd.read_csv(f'{dataset}_grid_val.csv')

label_col = f"{'partial' if results_to_optim == 'partial' else 'full'}_score"
other_label_col = f"{'full' if results_to_optim == 'partial' else 'partial'}_score"

df.drop(columns=["dataset", "n_epochs", "verbose", other_label_col],
        inplace=True)

if label_col == "partial_score":
    df[label_col] = df[label_col].apply(lambda x: float(x.split(" +- ")[0]))
df.head()
```

```python
# convert each categorical column to numeric and save the mapping
mappings = {}

for col in df.columns:
    if col != label_col:
        mapping = {label: idx for idx, label in enumerate(df[col].unique())}
        df[col] = df[col].map(mapping).astype(float)
        mappings[col] = mapping
```

## Parallel Coordinates Plot

```python
mappings['reg_type'].keys()
```

```python
import plotly.graph_objects as go
from matplotlib import cm

fig = go.Figure(
    go.Parcoords(
        line=dict(color=df[label_col], colorscale='viridis'),
        dimensions=[
            dict(label="reg_type", values=df["reg_type"], tickvals=list(mappings["reg_type"].values()),
                 ticktext=list(mappings["reg_type"].keys())),
            dict(label="weight_type", values=df["weight_type"], tickvals=list(mappings["weight_type"].values()),
                 ticktext=list(mappings["weight_type"].keys())),
            dict(label="alpha", values=df["alpha"], tickvals=list(mappings["alpha"].values()),
                 ticktext=list(mappings["alpha"].keys())),
            dict(label="lr", values=df["lr"], tickvals=list(mappings["lr"].values()),
                 ticktext=list(mappings["lr"].keys())),
            dict(label=f"{results_to_optim}_score", values=df[f"{results_to_optim}_score"]),
        ]
    )
)

# fig = px.parallel_coordinates(df, color=f"{results_to_optim}_score", color_continuous_scale=px.colors.diverging.Tealrose,)

fig.update_layout(
    plot_bgcolor='white',
    # paper_bgcolor = 'white'
)
fig.show()
```
